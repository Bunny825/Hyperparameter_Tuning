```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
data=sklearn.datasets.load_breast_cancer()
breast_data=pd.DataFrame(data.data,columns=data.feature_names)
```

```python
breast_data.head()
```

```python
breast_data["label"]=data.target
breast_data.head()
```

```python
breast_data.shape
```

```python
breast_data.isnull().sum()
```

```python
breast_data["label"].value_counts()
```

```python
x=breast_data.drop(columns="label",axis=1)
y=breast_data["label"]
```

```python
print(x)
print(y)
```

```python
x=np.asarray(x)
y=np.asarray(y)
```

#**GridSearchCV**



```python
model=SVC()
parameters={"kernel":["linear","poly","rbf","sigmoid"],"C":[1,5,10,20]}
classifier=GridSearchCV(model,parameters,cv=5)
classifier.fit(x,y)
print(classifier.best_params_)
print(classifier.best_score_)
```

```python
result=pd.DataFrame(classifier.cv_results_)
result.head()
grid_result=result[['param_C','param_kernel','mean_test_score']]
print(grid_result)
```

#**RandomizedSearchCV**

```python
model=SVC()
parameters={"kernel":["linear","poly","rbf","sigmoid"],"C":[1,5,10,20]}
classifier=RandomizedSearchCV(model,parameters,cv=5)
classifier.fit(x,y)
print(classifier.best_params_)
print(classifier.best_score_)
```
